#Chatbot
Proyecto de chatbot para respuestas básicas sobre el juego Valorant

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import datetime

print("Todo bien")
```

```python
# se debe anotar el diccionario de lucas ortega
banco_respuestas = {
    "hola": {"respuesta": "¡Hola! Bienvenido al chatbot ValoAcademy."},
}
```

```python
# Componentes de la interfaz visual dentro de Jupyter
#se deben agregar más cosas
salida_chat = widgets.Output()
entrada_texto = widgets.Text(placeholder='Escribe aquí tu consulta (ej: hola, horario, pagos)...', layout=widgets.Layout(width='70%'))
boton_enviar = widgets.Button(description='Enviar Consulta', button_style='primary')

def procesar_chat(b):
    with salida_chat:
        # Capturamos y normalizamos el texto ingresado
        mensaje = entrada_texto.value.lower().strip()
        entrada_texto.value = "" # Limpiamos la caja
        
        if not mensaje:
            return
            
        print(f"Usuario: {mensaje}")
        
        # Lógica de búsqueda indexada en el diccionario
        if mensaje in banco_respuestas:
            respuesta = banco_respuestas[mensaje]["respuesta"]
            print(f" Chatbot: -> {respuesta}\n")
        else:
            print("Chatbot: No comprendo tu consulta")

# Vinculamos el botón con la función lógica
boton_enviar.on_click(procesar_chat)

# Desplegamos la aplicación web simulada
print("💻 APLICACIÓN INTERACTIVA: CHATBOT DE SOPORTE")
display(widgets.HBox([entrada_texto, boton_enviar]), salida_chat)
```
